```python
asistentes={}
#Opción 'A' del 'Menú principal'
def agregar_asistente(asistentes, matricula, nombre, primer_apellido, segundo_apellido, fecha_nacimiento, carrera):
    asistentes[matricula] = {
        'nombre': nombre,
        'primer_apellido': primer_apellido,
        'segundo_apellido': segundo_apellido,
        'fecha_nacimiento': fecha_nacimiento,
        'carrera': carrera,
        'eventos_registrados': []
    }
    print(f"Asistente {nombre} {primer_apellido} agregado correctamente.")

def mostrar_asistentes(asistentes):
    if not asistentes:
        print("No hay asistentes registrados.")
        return

    print("\nLista de Asistentes:")
    for matricula, asistente in asistentes.items():
        print(f"Matrícula: {matricula}, Nombre: {asistente['nombre']} {asistente['primer_apellido']} "
              f"{asistente['segundo_apellido']} - Fecha de Nacimiento: {asistente['fecha_nacimiento']}, "
              f"Carrera: {asistente['carrera']}")

def actualizar_asistente(asistentes, matricula):
    if matricula in asistentes:
        print("\nSeleccione qué dato desea actualizar:")
        print("1. Nombre")
        print("2. Primer Apellido")
        print("3. Segundo Apellido")
        print("4. Fecha de Nacimiento")
        print("5. Carrera")
        print("6. Actualizar todos los datos")

        opcion = input("Seleccione una opción (1-6): ")

        if opcion in ['1', '2', '3', '4', '5']:
            nuevo_valor = input(f"Ingrese el nuevo {'nombre' if opcion == '1' else 'primer apellido' if opcion == '2' else 'segundo apellido' if opcion == '3' else 'fecha de nacimiento' if opcion == '4' else 'carrera'} del asistente: ")
            asistentes[matricula]['nombre' if opcion == '1' else 'primer_apellido' if opcion == '2' else 'segundo_apellido' if opcion == '3' else 'fecha_nacimiento' if opcion == '4' else 'carrera'] = nuevo_valor
        elif opcion == '6':
            nombre = input("Ingrese el nuevo nombre del asistente: ")
            primer_apellido = input("Ingrese el nuevo primer apellido del asistente: ")
            segundo_apellido = input("Ingrese el nuevo segundo apellido del asistente: ")
            fecha_nacimiento = input("Ingrese la nueva fecha de nacimiento del asistente (YYYY-MM-DD): ")
            carrera = input("Ingrese la nueva carrera del asistente(LA, LNI,CP,LTI,LGRS): ")

            asistentes[matricula] = {
                'nombre': nombre,
                'primer_apellido': primer_apellido,
                'segundo_apellido': segundo_apellido,
                'fecha_nacimiento': fecha_nacimiento,
                'carrera': carrera
            }

        print("Asistente actualizado correctamente.")
    else:
        print("Matrícula de asistente no encontrada.")

def eliminar_asistente(asistentes, matricula):
    if matricula in asistentes:
        asistente = asistentes.pop(matricula)
        print(f"Asistente {asistente['nombre']} {asistente['primer_apellido']} eliminado correctamente.")
    else:
        print("Matrícula de asistente no encontrada.")

def menu(asistentes):
    while True:
        print("\n--- Menú de Registro ---")
        print("1. Agregar Asistente")
        print("2. Mostrar Asistentes")
        print("3. Actualizar Asistente")
        print("4. Eliminar Asistente")
        print("5. Salir")

        opcion = input("Seleccione una opción (1-5): ")

        if opcion == '1':
            matricula = (input("Ingrese la matrícula del asistente: "))
            nombre = input("Ingrese el nombre del asistente: ")
            primer_apellido = input("Ingrese el primer apellido del asistente: ")
            segundo_apellido = input("Ingrese el segundo apellido del asistente: ")
            fecha_nacimiento = input("Ingrese la fecha de nacimiento del asistente (YYYY-MM-DD): ")
            carrera = input("Ingrese la carrera del asistente (LA, LNI, CP, LTI, LGRS): ")
            agregar_asistente(asistentes, matricula, nombre, primer_apellido, segundo_apellido, fecha_nacimiento, carrera)

        elif opcion == '2':
            mostrar_asistentes(asistentes)

        elif opcion == '3':
            matricula = (int(input("Ingrese la matrícula del asistente a actualizar: ")))
            actualizar_asistente(asistentes, matricula)

        elif opcion == '4':
            matricula = (int(input("Ingrese la matrícula del asistente a eliminar: ")))
            eliminar_asistente(asistentes, matricula)

        elif opcion == '5':
            print("Saliendo del Menú asistente.")
            break

        else:
            print("Opción no válida. Por favor, seleccione una opción válida.")


#Opción 'B' del 'Menú principal'
def registrar_asistente_evento(asistentes, matricula_buscada):
    if matricula_buscada in asistentes:
        asistente = asistentes[matricula_buscada]

        print("Selecciona el evento al que deseas registrar al asistente:")
        print("[A] Inteligencia Artificial en los Negocios")
        print("[B] Uso de la nube para gestión de procesos")
        print("[C] Industria 4.0 retos y oportunidades")
        print("[D] Machine Learning for a better world")
        print("[E] Retos de la Banca Electrónica en México")

        while True:
            evento_opcion = input("Opción: ").upper()

            if evento_opcion in ['A', 'B', 'C', 'D', 'E']:
                evento = eventos[evento_opcion]
                if len(evento['asistentes']) < evento['capacidad']:
                    evento['asistentes'].append(asistente)
                    asistente['eventos_registrados'].append(evento)
                    print(f"Asistente registrado en el evento {evento['nombre']}.")
                    break
                else:
                    print("El evento está lleno. No se puede registrar al asistente.")
            else:
                print("Opción no válida. Inténtalo de nuevo.")
    else:
        print("Asistente no encontrado.")

#Opción 'C' del 'Menú principal'
def marcar_asistencia(asistentes, matricula_buscada):
    if matricula_buscada in asistentes:
        asistente = asistentes[matricula_buscada]

        print("Selecciona el evento al que deseas marcar asistencia:")
        for evento in asistente['eventos_registrados']:
            print(f"[{evento['codigo']}] {evento['nombre']} - {evento['fecha']} {evento['hora']}")

        while True:
            evento_opcion = input("Opción: ").upper()

            evento = next((e for e in eventos.values() if e['codigo'] == evento_opcion), None)

            if evento:
                evento['asistencia'].append(asistente)
                print(f"Asistencia registrada para el evento {evento['nombre']}.")
                break
            else:
                print("Evento no encontrado. Inténtalo de nuevo.")
    else:
        print("Asistente no encontrado.")

#Opción 'D' del 'Menú principal'
def ver_eventos_alumno(asistentes, matricula_buscada):
    if matricula_buscada in asistentes:
        asistente = asistentes[matricula_buscada]

        print("Eventos registrados por el asistente:")
        for evento in asistente['eventos_registrados']:
            print(f"- {evento['nombre']} - {evento['fecha']} {evento['hora']}")
    else:
        print("Asistente no encontrado.")

#Opción 'E' del 'Menú principal'
def listado_asistencias(eventos):
    print("Listado de asistencias:")
    for evento in eventos.values():
        print(f"{evento['nombre']} ({len(evento['asistencia'])}/{evento['capacidad']} asistentes)")

eventos = {
    'A': {'codigo': 'A', 'nombre': 'Inteligencia Artificial en los Negocios', 'fecha': '4 de noviembre', 'hora': '15:00', 'capacidad': 1000, 'asistentes': [], 'asistencia': []},
    'B': {'codigo': 'B', 'nombre': 'Uso de la nube para gestión de procesos', 'fecha': '5 de noviembre', 'hora': '9:00', 'capacidad': 200, 'asistentes': [], 'asistencia': []},
    'C': {'codigo': 'C', 'nombre': 'Industria 4.0 retos y oportunidades', 'fecha': '5 de noviembre', 'hora': '14:00', 'capacidad': 150, 'asistentes': [], 'asistencia': []},
    'D': {'codigo': 'D', 'nombre': 'Machine Learning for a better world', 'fecha': '5 de noviembre', 'hora': '19:00', 'capacidad': 150, 'asistentes': [], 'asistencia': []},
    'E': {'codigo': 'E', 'nombre': 'Retos de la Banca Electrónica en México', 'fecha': '6 de noviembre', 'hora': '15:00', 'capacidad': 1000, 'asistentes': [], 'asistencia': []},
}

#Opción 'F' del 'Menú principal'
def emitir_constancia(asistentes, matricula):
    asistente = asistentes.get(matricula)
    if asistente:
        eventos_registrados = asistente.get('eventos_registrados', [])
        if len(eventos_registrados) >= 3:
            print(f"Constancia para el alumno {asistente['nombre']} {asistente['primer_apellido']} por asistir a 3 o más eventos.")
        else:
            print(f"El alumno {asistente['nombre']} {asistente['primer_apellido']} no ha asistido a suficientes eventos para recibir la constancia.")
    else:
        print("Asistente no encontrado.")

```

```python
#Validaciones para la opción escogida en el menú principal
import re
def elegir_letra(prompt:str='Dame una letra: ',
                 opciones_validas:str='12345'):
        while True:
            opcion=input(prompt)
            opcion=opcion.upper()
            if (opcion==''):
                print('Error en captura. Opción no se puede omitir. Inténtelo de nuevo')
                continue
            if not bool(re.match(f'^[{opciones_validas}]$',opcion)):
                print('Error en captura. Opción no reconocida. Inténtelo de nuevo')
                continue
            # Si todo fue bien, se sale.
            break
        return opcion
```

```python
def mostrar_menu(
    opciones:dict,
    titulo:str='OPCIONES DISPONIBLES'):

    print(titulo)
    opciones_validas=''
    for k,v in opciones.items():
        print(f'[{k}] {v}')
        opciones_validas=f'{opciones_validas}{k}'

    opc=elegir_letra('Qué opción deseas?: ', opciones_validas)
    return opc
```

```python
4# Programa principal

opciones_menu_principal={
    'A':'Registrar un asistente',
    'B':'Registrar asistente a un evento',
    'C':'Registrar asistencia al evento',
    'D':'Ver eventos del alumno',
    'E':'Listado de asistencia',
    'F': 'Constancia de participación',
    'X':'Salir\n'
}

opciones_menu_asistentes={
    'A':'Registrar un asistente',
    'B':'Eliminar asistente',
    'C':'Modificar datos de asistente',
    'D':'Consultar asistente',
    'X':'Salir\n'
}

while True:
    opcion_elegida=mostrar_menu(opciones_menu_principal, '\n** MENÚ PRINCIPAL\n')
    match opcion_elegida:
        case 'A':
            menu(asistentes)
        case 'B':
          matricula_buscada = str(input("Matrícula del asistente: "))
          registrar_asistente_evento(asistentes, matricula_buscada)
        case 'C':
          matricula_buscada = str(input("Matrícula del asistente: "))
          marcar_asistencia(asistentes,matricula_buscada)
        case 'D':
          matricula_buscada = str(input("Matrícula del asistente: "))
          ver_eventos_alumno(asistentes,matricula_buscada)
        case 'E':
          listado_asistencias(eventos)
        case 'F':
          matricula = str(input("Matrícula del asistente: "))
          emitir_constancia(asistentes,matricula)
        case 'X':
            print('Gracias por usar el sistema.')
            break
        case _:
            print('Opción no reconocida.')
            continue

```
